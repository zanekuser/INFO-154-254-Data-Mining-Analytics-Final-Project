```python
import pandas as pd
```

```python
long_file = 'Medicare Provider Utilization and Payment Data_Physician and Other Supplier PUF_CY2016.csv'
short_file = 'Medicare Physician and Other Supplier NPI Aggregate Report_CY2016.csv'
path = 'data/'
short = pd.read_csv(path+short_file)
long = pd.read_csv(path+long_file)
```

```python
long.columns
```

```python
grouped = (long.loc[:,['HCPCS Code','HCPCS Description','Average Submitted Charge Amount']]
    .groupby(['HCPCS Code','HCPCS Description']).agg(
        {'Average Submitted Charge Amount': 'std',
        'HCPCS Code': 'count'}))
grouped = grouped.rename({'Average Submitted Charge Amount': 'charge_std', 'HCPCS Code': 'count'}, axis = 1)
grouped.sort_values('charge_std',ascending = False)
```
